```python
import pystar as ps
import numpy as np
ps.set_log_level("INFO")
```

```python
config = ps.pystar.StarConfig()
config.chuck_size = 1024*1024*1
config.compression = ps.CompressionAlgorithm.LZ4

ds = ps.StarDataset.create("/tmp/test.star", config=config) 
r = 400000
c = 6 
ds["data"]  = np.arange(r*c).reshape(r, c).astype(np.float64)
ds["test"] = ["5"]
ds.meta["name"] = "instrument"
layer1 = ds.create_layer("layer_1"); 
layer1["data"] = np.arange(100).astype(np.float64);
layer1.meta["focal"] = 10.5 
ds.meta["test"]="MRO"
ds.meta["data1"] = [1,2,4,5]
print("Done")
ds.close()
```
